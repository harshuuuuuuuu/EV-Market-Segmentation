# Market Segmentation & Analysis Of EV Market
Automobiles that run on electric motors rather than internal combustion engines (ICEs), which depend on
fossil fuels, are known as electric vehicles (EVs). They store electricity and supply it to the electric motor that
drives the car using rechargeable batteries or other energy storage technologies. Compared to conventional
ICE vehicles, EVs have a number of advantages, making them a more and more common choice among
consumers and a crucial part of sustainable transportation solutions.
A major obstacle to the attainment of sustainable development goals is posed by the use of electric vehicles,
which offer a potential solution to the rising levels of global air pollution. Due to India's urgent need for
improved infrastructure and more benevolent governmental laws, electric vehicles (EVs) have become an
essential part of the country's energy and mobility sectors.

- Demographic Analysis: This section includes visualizations and insights based on variables like age, profession, marital status, education, salary, and personal loan status.
- Market Segmentation Analysis: Focuses on segmenting the market based on various factors such as profession, age, marital status, education, total salary, EV make, and personal loan.

# K-Means Modelling
- K-Means Clustering: This section involves applying the K-Means clustering algorithm to segment the EV car dataset into clusters based on selected features.

- Cluster Visualization: Visualizes the clusters using PCA for dimensionality reduction to help understand the grouping of EV cars in the dataset.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA


features = df2[['Battery Capacity(kWh)', 'Acceleration(sec)', 'TopSpeed(km/h)', 'Range(km)', 'Price(Lh)']]

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), inertia, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()

optimal_k = 3  
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
cluster_labels = kmeans.fit_predict(scaled_features)

df2['Cluster'] = cluster_labels

pca = PCA(n_components=2)
reduced_features = pca.fit_transform(scaled_features)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=reduced_features[:, 0], y=reduced_features[:, 1], hue=cluster_labels, palette='Set1', s=100)
plt.title('Clusters of EV Cars (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster', loc='upper right')
plt.show()
```
